```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('udemy_courses.csv')
```

```python
data.head()
```

```python
# Convert published_timestamp to datetime format
data.dtypes
```

```python
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])
```

```python
data.dtypes
```

```python
data.head()
```

```python
# Shape of dataset
print("Number of rows", data.shape[0])
print("Number of columns", data.shape[1])
```

```python
# Info about dataset
data.info()
```

```python
# Check null values in dataset
data.isnull().sum()
```

```python
sns.heatmap(data.isnull()) # No any missing value in the dataset
```

```python
# Check for duplicate data 
dup=data.duplicated().any()
print("Are there any duplicated value in the dataset ", dup)
```

```python
# Drop duplicates data
data.drop_duplicates(inplace = True)
```

```python
dup=data.duplicated().any()
print("Are there any duplicated value in the dataset ", dup)
```

```python
data.head()
```

```python
# Number of courses per subject
data['subject'].value_counts()
```

```python
sns.countplot(data['subject'])
plt.xticks(rotation=90)
plt.xlabel('Subjects',fontsize=13)
plt.ylabel('Number of courses per subject',fontsize=13)
```

```python
# For which levels, udemy courses providing the courses
data['level'].unique()
```

```python
data['level'].value_counts()
```

```python
sns.countplot(data['level'])
plt.xticks(rotation=90)
plt.xlabel('Level',fontsize=13)
plt.ylabel('Number of courses per Level',fontsize=13)
```

```python
# Display the count of paid and free courses
data['is_paid'].value_counts()
```

```python
sns.countplot(data['is_paid'])
plt.xticks(rotation=90)
plt.xlabel('Paid',fontsize=13)
plt.ylabel('Number of Free and Paid courses',fontsize=13)
```

```python
# Which courses has more lectures(free or paid)?
data.groupby(['is_paid'])['num_lectures'].mean()
```

```python
# Which courses have a higher number of subscribers free or paid?
data.groupby(['is_paid'])['num_subscribers'].mean()
```

```python
sns.barplot(x='is_paid', y='num_subscribers', data=data)
```

```python
# Which level has the highest number of subscribers?
data.groupby(['level'])['num_subscribers'].mean()
```

```python
sns.barplot(x='level', y='num_subscribers', data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
# Find most polular course title
data[data['num_subscribers'].max() == data['num_subscribers']]['course_title']
```

```python
# Display 10 most popular course as per number of subscribers
top_10 = data.sort_values(by='num_subscribers', ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers', y='course_title', data=top_10)
plt.xlabel("Number of subscribers")
plt.ylabel("Course Title")
plt.show()
```

```python
# Find the course wich is having the highest number of reviews
data[data['num_reviews'].max() == data['num_reviews']]
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='subject', y='num_reviews', data=data)
```

```python
# Does price affect number of reviews?
plt.figure(figsize=(15,6))
sns.scatterplot(x='price', y='num_reviews', data=data)
```

```python
# Find the total number of courses related to python
len(data[data['course_title'].str.contains('python', case=False)])
```

```python
# Display 10 most popular courses as per number of subscribers
python = data[data['course_title'].str.contains('python', case=False)].sort_values(by='num_subscribers', ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers', y='course_title', data=python)
```
